```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Ice Cream Sales - temperatures.csv')
data
```

```python
data.info()
```

```python
# @title Temperature vs Ice Cream Profits

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Temperature']
  ys = series['Ice Cream Profits']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = data.sort_values('Temperature', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Temperature')
_ = plt.ylabel('Ice Cream Profits')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.title('The Effect of Temperature on Ice Cream Sales')
sns.lineplot(x=data['Temperature'], y=data['Ice Cream Profits'])
plt.show()
```
